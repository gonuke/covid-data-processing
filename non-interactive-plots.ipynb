```python
import numpy as np
import pandas as pd
import plots
plots.use_JHU_github_live_data()
```

```python
country_list = ["US","Canada","Israel","Italy","Spain","Sweden","United Kingdom"]
fit_info = {'constant':2, 'length':5, 'type':'exp'}
```

```python
cases, deaths, recovered = plots.parse_country_data()
```

```python
for data_set, data_type, threshold in zip([deaths,cases],["deaths","cases"],[5,100]):
    plots.semilog_since(data_set, country_list, data_type=data_type,
                                threshold=threshold,
                                fit_info = fit_info)
```

```python
for data_set, data_type, threshold in zip([cases,deaths],["cases","deaths"],[1,1]):
    plots.semilog_per_capita_since(data_set, country_list, 
                                   data_type=data_type,
                                   threshold=threshold,
                                   fit_info = fit_info)
```
